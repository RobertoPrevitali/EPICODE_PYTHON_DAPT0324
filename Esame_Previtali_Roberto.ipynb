```python
'''DOMANDA 2): Si chiede poi per ogni continente di trovare il numero di casi totali avvenuti in quello stesso continente; si chiede di non considerare eventuali locazioni che nel dataset non appartengono ad alcun continente;'''
'''RISPOSTA 2): Analizzando il DataFrame ho notato che le righe con continent=Nan hanno come 
location il nome di un continente, tali righe rappresentano la subtotali dei dati di ogni 
continente. Posso raggruppare per continenti e sommare i new_cases usando la colonna continent'''
print(cov.groupby(["continent"]).sum("new_cases")["new_cases"])
print(cov.groupby(["location"]).max("total_cases")["total_cases"]) #Questo comando per evidenziare che la somma dei new_cases per continent ed il valore max di casi_totali dei subtotali sono molto simili (VEDI Africa)
```

```python
'''RISPOSTA 3): in virtù delle precedenti verifiche, ho interpretato la domanda in questo modo:
A: essendo casi_totali incrementale, ha senso cercare il valore massimo tra le singole nazioni
B: la media richiesta dei casi totali è una media dei casi totali tra le nazioni di ogni continente
C: la richiesta "(calcolati anche sulle locazioni senza indicazione di continente)" è relativa
solo alla percentuale rispetto al numero di casi totali nel mondo, quindi utilizzerò il subtotale World
che tiene conto di tutte le location'''

cont1="Asia"
cont2="Europe"
tot_world=cov.loc[(cov["location"]=="World") & (cov["date"]==cov.iloc[-1,3])]["total_cases"] #cerco i casi totali nel mondo

#Primo continente
print("Primo continente:",cont1)
filtro_cont1=cov.loc[cov["continent"]==cont1]
max_tot_cases1=filtro_cont1.groupby(["continent","location"]).max("total_cases")
print("Massimo per ogni nazione")
print(max_tot_cases1["total_cases"]) #trovo il massimo per tutte le singole nazioni
print("\n")
max_tot_cases1_overall=max_tot_cases1.sort_values("total_cases", ascending=False)
print("La nazione con più casi totali in", cont1,"è: ", max_tot_cases1_overall.iloc[0,0:1]) #ottengo la nazione con i maggiori casi totali
print("\n")
print("Il valore medio dei casi totali per nazione in", cont1,"è: ", round(max_tot_cases1["total_cases"].mean(),0)) #ottengo la media tra tutte le nazioni
print("\n")
print("La percentuale dei casi totali in", cont1,"rispetto ai casi totali nel mondo è: ",round(100*max_tot_cases1["total_cases"].sum()/tot_world,2))
print("\n")
print("\n")

#Secondo continente
print("Secondo continente:",cont2)
filtro_cont2=cov.loc[cov["continent"]==cont2]
max_tot_cases2=filtro_cont2.groupby(["continent","location"]).max("total_cases")
print("Massimo per ogni nazione")
print(max_tot_cases2["total_cases"]) #trovo il massimo per tutte le singole nazioni
print("\n")
max_tot_cases2_overall=max_tot_cases2.sort_values("total_cases", ascending=False)
print("La nazione con più casi totali in", cont2,"è: ", max_tot_cases2_overall.iloc[0,0:1]) #ottengo la nazione con i maggiori casi totali
print("\n")
print("Il valore medio dei casi totali per nazione in", cont2,"è: ", round(max_tot_cases2["total_cases"].mean(),0)) #ottengo la media tra tutte le nazioni
print("\n")
print("La percentuale dei casi totali in", cont2,"rispetto ai casi totali nel mondo è: ",round(100*max_tot_cases2["total_cases"].sum()/tot_world,2))
print("\n")

#Ho volutamente lasciato le nazioni prive di dati (es. England, Hong Kong), per evidenziare la mancanza di dati.
#La media non viene viziata da tali mancanze
```

```python
'''DOMANDA 4):Selezionare i dati relativi all'Italia nel 2022, e mostrare con un grafico adeguato 
l'evoluzione dei casi totali rispetto alla data. Mostrare poi con un grafico adeguato il numero di
nuovi casi rispetto alla data (filtrare i dati se necessario). Mostrare infine un grafico che 
mostra l'andamento della somma cumulativa nuovi casi del 2022, e commentare se ci sono similarità 
con l'andamento del numero dei casi totali'''


'''RISPOSTA 4):'''
#EVOLUZIONE DATI
Italy2022=cov.loc[(cov["location"]=="Italy")&(cov["date"]>="2022-01-01")&(cov["date"]<="2022-12-31")]
Italy2022.plot(x="date", y="total_cases", xlabel="Date", ylabel="Total cases",title="Italy's total cases evolution (x10millions)",figsize=[12,12])

#NUOVI CASI (eliminati i giorni con nuovi casi=0)
Italy2022.loc[Italy2022["new_cases"]>0].plot(kind="bar", x="date", y="new_cases", xlabel="Date", ylabel="New cases",title="Italy's new cases (millions)", figsize=[12,12])

#SOMMA CUMULATA NUOVI CASI
Italy2022["cumulative_new_cases"]=Italy2022["new_cases"].cumsum() #creo nuova colonna
Italy2022.plot(x="date", y="cumulative_new_cases", xlabel="Date", ylabel="Cumulative new cases",title="Italy's cumulative cases (x10millions)",figsize=[12,12])

#COMPARAZIONE CON ANDAMENTO TOTAL_CASES
Italy2022.plot(x="date", y=["total_cases","cumulative_new_cases"], xlabel="Date", ylabel="Cases",title="Italy's Total and Cumulative cases (x10millions)",figsize=[12,12])
'''Guardando il grafico di comparazione, l'andamento dei casi totali e dei nuovi casi risulta similare'''
```

```python
'''DOMANDA 5):Riguardo le nazioni di Italia, Germania e Francia, mostrare in un boxplot la differenza 
tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU) 
da maggio 2022 (incluso) ad aprile 2023 (incluso), e scrivere un breve commento a riguardo'''

'''RISPOSTA 5)'''
Italy0522_0423=cov.loc[(cov["location"]=="Italy")&(cov["date"]>="2022-05-01")&(cov["date"]<="2023-04-30")]
France0522_0423=cov.loc[(cov["location"]=="France")&(cov["date"]>="2022-05-01")&(cov["date"]<="2023-04-30")]
Germany0522_0423=cov.loc[(cov["location"]=="Germany")&(cov["date"]>="2022-05-01")&(cov["date"]<="2023-04-30")]
Ita_Fra_Ger_0522_0423=pd.concat([Italy0522_0423,France0522_0423,Germany0522_0423])
sns.boxplot(data=Ita_Fra_Ger_0522_0423,x="location",y="icu_patients").set(xlabel="Country", ylabel="Number of ICU Patients", title="ICU Patients")

'''Nel periodo studiato, l'Italia ha avuto un numero di pazienti in terapia intensiva nettamente
inferiore rispetto a Francia e Germania. La mediana di Francia e Germania è molto simile ma le 
curtosi risultano diverse: nello specifico la distribuzione dei pazienti ICO in Germania 
risulta più piatta rispetto alla distribuzione dei pazienti ICO francese'''
```

```python
''' DOMANDA 6): Riguardo le nazioni di Italia, Germania, Francia e Spagna nel 2023, mostrare 
la somma dei pazienti ospitalizzati per ognuna - se ci sono dati nulli, suggerire se può essere
possibile gestirli tramite sostituzione o meno'''

'''RISPOSTA 6): usare il dato hosp_patients per sommare i pazienti ospitalizzati
nell'anno 2023 non sembra corretto in quando hosp_patients è un valore spot può tenere conto di 
chi viene ospitalizzato e chi viene dimesso. Il dato weekly_hosp_admissions sembra essere più 
coerente con la richiesta, essendo però un dato rolling di 7 giorni (giorno stesso + 6gg precedenti)
bisogna prima fare uno slicing dei dati in modo tale da sommare correttamente i nuovi ingressi. 
Per quanto riguarda la presenza di dati nulli più che sostituire i dati forse è più 
opportuno: 
A) considerare i dati presenti
B) fare un ulteriore filtro temporale per il quale tutte le serie presentano dati, ovvero
il primo trimestre'''

#filtro i dati per anno 2023 e considero solo un giorno ogni settimana
Italy2023=cov.loc[(cov["location"]=="Italy")&(cov["date"]>="2023-01-01")&(cov["date"]<="2023-12-31")]
Germany2023=cov.loc[(cov["location"]=="Germany")&(cov["date"]>="2023-01-01")&(cov["date"]<="2023-12-31")]
France2023=cov.loc[(cov["location"]=="France")&(cov["date"]>="2023-01-01")&(cov["date"]<="2023-12-31")]
Spain2023=cov.loc[(cov["location"]=="Spain")&(cov["date"]>="2023-01-01")&(cov["date"]<="2023-12-31")]
weekly_Italy2023=Italy2023.iloc[::7,:]
weekly_Germany2023=Germany2023.iloc[::7,:]
weekly_France2023=France2023.iloc[::7,:]
weekly_Spain2023=Spain2023.iloc[::7,:]
weekly_Italy_Germany_France_Spain2023=pd.concat([weekly_Italy2023,weekly_Germany2023,weekly_France2023,weekly_Spain2023])
print("Dati ospitalizzazioni su tutto il 2023")
print(weekly_Italy_Germany_France_Spain2023.groupby("location").sum("weekly_hosp_admissions")["weekly_hosp_admissions"])
print("\n")

#filtro i dati per il primo quadrimestre anno 2023 e considero solo un giorno ogni settimana
Italy20231q=cov.loc[(cov["location"]=="Italy")&(cov["date"]>="2023-01-01")&(cov["date"]<="2023-03-31")]
Germany20231q=cov.loc[(cov["location"]=="Germany")&(cov["date"]>="2023-01-01")&(cov["date"]<="2023-03-31")]
France20231q=cov.loc[(cov["location"]=="France")&(cov["date"]>="2023-01-01")&(cov["date"]<="2023-03-31")]
Spain20231q=cov.loc[(cov["location"]=="Spain")&(cov["date"]>="2023-01-01")&(cov["date"]<="2023-03-31")]
weekly_Italy20231q=Italy20231q.iloc[::7,:]
weekly_Germany20231q=Germany20231q.iloc[::7,:]
weekly_France20231q=France20231q.iloc[::7,:]
weekly_Spain20231q=Spain20231q.iloc[::7,:]
weekly_Italy_Germany_France_Spain20231q=pd.concat([weekly_Italy20231q,weekly_Germany20231q,weekly_France20231q,weekly_Spain20231q])
print("Dati ospitalizzazioni sul primo trimestre 2023") 
print(weekly_Italy_Germany_France_Spain20231q.groupby("location").sum("weekly_hosp_admissions")["weekly_hosp_admissions"])
```
